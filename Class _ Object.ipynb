```python
class Account():
  def __init__(self, name="blank"):
    self.amount = 0
    self.name = name

  def deposit(self, money):
    self.amount += money
    print(f"{self.name}님의 계좌에 {money}원이 입금 되었습니다")
```

```python
object1 = Account("이선홍")
```

```python
object1.deposit(5000)
```

```python
object2 = Account("성균관대학교")
```

```python
object2.deposit(5000000)
```

```python
class Revised_Account(Account):
    def __init__(self, name="blank"):
      super().__init__(name)

    def print_amount(self):
      print(f"현재 {self.name}님은 계좌에 {self.amount}원 있습니다")

    def withdraw(self, money):
      if self.amount - money < 0:
        print("돈이 없습니다..")
      else:
        self.amount -= money
        print(f"{self.name}님의 계좌에는 {money}원을 출금하여 {self.amount}원이 남았습니다.") 
```

```python
object3 = Revised_Account("이선홍")
```

```python
object3.deposit(50000)
object3.print_amount()
```

```python
object3.deposit(50000)
object3.print_amount()
```

```python
object3.withdraw(40000)
object3.withdraw(50000)
object3.withdraw(50000)
object3.print_amount()
```
